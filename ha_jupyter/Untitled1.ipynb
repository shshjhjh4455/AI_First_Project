```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
df = pd.read_csv('C:/Users/hyha1/Documents/GitHub/AI_First_Project/customer_payment_data.csv', encoding='cp949')
```

```python
df.head(10)
```

```python
# date 변수에 df['날짜'] 변수들 저장
date = df['날짜']
date
```

```python
# '전국'이라는 단어가 들어간 컬럼 추출
df = df[df.columns[pd.Series(df.columns).str.startswith('전국')]]
df.head()
```

```python
# 날짜 변수 추가이후 순서변경을 위한 변수명 리스트에 '날짜' 추가.
col = df.columns
col = col.insert(0, '날짜')
col
```

```python
# df['날짜']에 date 변수 추가(date 변수에는 날짜가 들어있음)
df['날짜'] = date
```

```python
# [날짜]가 끝에 추가된 것 확인
df.head()
```

```python
# df에 변수 [col]에 넣어둔 순서를 입력 
df = df[col]
```

```python
# 변경된 날짜 위치 확인
df
```

```python
covid_2017 = df[df['날짜'].between('2017/01', '2017/12')]
covid_2017
```

```python
covid_2017.set_index('날짜', inplace=True)
covid_2017
```

```python
plt.figure(figsize = (25,9))
plt.plot(covid_2017['전국_합계'])
plt.show()
```
