```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#데이터 불러오기
df = pd.read_csv('C:/Users/hyha1/Documents/GitHub/AI_First_Project/customer_payment_data.csv', encoding='cp949')
```

```python
def region_extract(df, region:str):
    date = df['날짜']
    col = df.columns
    
    region_col = [c for c in col if c[:2] == region]
#    print(region_col)
    region_df = df[region_col]
    region_df.insert(0, '날짜', date)
#    print(region_df.head())
    return region_df
```

```python
## 2017년부터 자료 추출 

for i in range(len(df)):
    if df['날짜'][i] == '2017/01':
        df = df[i:]
df
```

```python
r_df = region_extract(df, '서울')
r_df.head()
```

```python
month_2017 = r_df[r_df['날짜'].between('2017/01', '2017/12')]
month_2018 = r_df[r_df['날짜'].between('2018/01', '2018/12')]
month_2019 = r_df[r_df['날짜'].between('2019/01', '2019/12')]
month_2020 = r_df[r_df['날짜'].between('2020/01', '2020/12')]
month_2021 = r_df[r_df['날짜'].between('2021/01', '2021/12')]
month_2022 = r_df[r_df['날짜'].between('2022/01', '2022/12')]
```

```python
month_2017.head()
```

```python
month_2017.set_index('날짜', inplace=True)
month_2018.set_index('날짜', inplace=True)
month_2019.set_index('날짜', inplace=True)
month_2020.set_index('날짜', inplace=True)
month_2021.set_index('날짜', inplace=True)
month_2022.set_index('날짜', inplace=True)
```

```python
month_2022
```

```python
x_2017 = month_2017.sum()
x_2018 = month_2018.sum()
x_2019 = month_2019.sum()
x_2020 = month_2020.sum()
x_2021 = month_2021.sum()
x_2022 = month_2022.sum()
```

```python
sum_df = pd.DataFrame(pd.concat([pd.DataFrame(x_2017).T, pd.DataFrame(x_2018).T, pd.DataFrame(x_2019).T, pd.DataFrame(x_2020).T, 
           pd.DataFrame(x_2021).T, pd.DataFrame(x_2022).T], axis=0))

sum_df
```

```python
sum_df['year'] = ['2017','2018','2019','2020','2021','2022']
```

```python
sum_df=sum_df.set_index('year')
```
