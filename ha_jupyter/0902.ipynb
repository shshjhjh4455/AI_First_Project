```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
df = pd.read_csv('C:/Users/hyha1/Documents/GitHub/AI_First_Project/customer_payment_data.csv', encoding='cp949')
```

```python
df.head(10)
```

```python
df.info()
```

```python
# date 변수에 df['날짜'] 변수들 저장
date = df['날짜']
date
```

```python
# '전국'이라는 단어가 들어간 컬럼 추출
df = df[df.columns[pd.Series(df.columns).str.startswith('전국')]]
df.head()
```

```python
# 날짜 변수 추가이후 순서변경을 위한 변수명 리스트에 '날짜' 추가.
col = df.columns
col = col.insert(0, '날짜')
col
```

```python
# df['날짜']에 date 변수 추가(date 변수에는 날짜가 들어있음)
df['날짜'] = date
```

```python
# [날짜]가 끝에 추가된 것 확인
df.head()
```

```python
# df에 변수 [col]에 넣어둔 순서를 입력 
df = df[col]
```

```python
# 변경된 날짜 위치 확인
df
```

```python
# 코로나 전 데이터 , between 사용해서 날짜 기준 으로 지정
covid_bef = df[df['날짜'].between('2017/07', '2019/12')]
covid_bef
```

```python
# 시각화를 위해 index에 '날짜' 지정
covid_bef.index = covid_bef['날짜']
covid_bef.head()
```

```python
# 날짜 컬럼을 제외한 컬럼 조회
covid_bef = covid_bef[covid_bef.columns.difference(['날짜'])]
covid_bef
```

```python
# plot 를 사용하여 그래프 생성  
plt.figure(figsize = (25,9))
plt.plot(covid_bef['전국_합계'])
plt.show()
```

```python
covid_bef.info()
```

```python
# 코로나 후, between 사용해서 기간 설정  
covid_aft = df[df['날짜'].between('2020/01', '2022/05')]
covid_aft.head()
```

```python
# 시각화를 위해 날짜 인덱스 지정 및 칼럼에서 날짜 제외한 데이터를 조회 
covid_aft.index = covid_aft['날짜']
covid_aft = covid_aft[covid_aft.columns.difference(['날짜'])]
covid_aft.head()
```

```python
# plot 를 사용하여 '전국_합계' 컬럼 기준으로 시각화
plt.figure(figsize = (25,9))
plt.plot(covid_aft['전국_합계'])
plt.show()
```

```python
# 코로나 전과 후 데이터 concat 사용하여 합침 
all_df = pd.concat([covid_bef, covid_aft], axis=0)
all_df.head()
all_df.to_csv("update_customer_payment_data.csv", index = False, encoding = 'utf-8-sig')
```

```python
# 끝 5개 데이터 조회
all_df.tail()
```

```python
# market 는 *을 기준으로, 사이즈 10을 줌, 전체 기간 중 코로나 이전은 green으로 설정 후 시각화
plt.figure(figsize = (25,9))
plt.plot(all_df['전국_합계'], marker = '*', markersize =10)
plt.plot(all_df['2017/07' : '2019:12']['전국_합계'], color='green', marker = '*', markersize =10)
plt.show()
```

```python
#카테고리별 코로나 전후로 평균값을 구해 소수점 두번째자리까지 지정 ,  코로나 전 카테고리 평균값

category_mean_bef = pd.DataFrame()
category_mean_aft = pd.DataFrame()
for c in col:
    bef_mean = round(covid_bef[c].mean(),2)
    aft_mean = round(covid_aft[c].mean(),2)

    category_mean_bef[f'bef_{c}_mean'] = [bef_mean]
    category_mean_aft[f'aft_{c}_mean'] = [aft_mean]

category_mean_bef
```

```python
#코로나 후 카테고리 평균값
category_mean_aft
```

```python
# 카테고리 전과 후 칼럼별로 막대그래프로 비교 

for b,a in zip(category_mean_bef.columns, category_mean_aft.columns):
    plot_df = pd.concat([category_mean_bef[[b]], category_mean_aft[[a]]], axis=0)
    plot_df.plot(y=[b,a], kind="bar",figsize=(3,5))
#     plt.xticks()
    plt.legend(loc="lower center", bbox_to_anchor=(1,1))
    plt.show()
```

```python
#그래프를 새로 구하기 위한 방법 
new_index = all_df.columns
new_index
```

```python
# after - before 
# after에서 before를 뺀 값이 0보다 크면, covid_df_true 에 넣은 후, covid_df_true에 새로운 칼럼 ['after-before']을 추가해서
# 차이값도 확인
covid_df_true = covid_df[(covid_df['after'] - covid_df['before'] > 0)]
covid_df_true['after-before'] = covid_df_true['after'] - covid_df_true['before']
```

```python
##코로나 이후 소비량이 늘어난 목록들 

print(len(covid_df_true))
covid_df_true
```

```python
# before - after
# 코로나 이후 소비량이 줄어든 목록들
covid_df_false = covid_df[(covid_df['before'] - covid_df['after'] > 0)]
covid_df_false['before-after'] = covid_df_false['before'] - covid_df_false['after']
```

```python
#코로나 이후 소비량이 줄어든 목록 갯수 및 카테고리
print(len(covid_df_false))
covid_df_false
```

```python
#새로운 데이터프레임 추가(증감율 보기 위해)
mean_covid_df = pd.DataFrame({'before': covid_bef_mean, 'after':covid_aft_mean}, 
                            index =new_index)
mean_covid_df.head()
```

```python
# 증감률 = (최종수치 - 최초수치)/최초수치 *100
mean_covid_df['증감률'] = (mean_covid_df['after']-mean_covid_df['before'])/mean_covid_df['before'] * 100
mean_covid_df.head()
```
