```python
#데이터 불러오기
df = pd.read_csv('C:/Users/hyha1/Documents/GitHub/AI_First_Project/customer_payment_data.csv', encoding='cp949')
```

```python
def region_extract(df, region:str):
    date = df['날짜']
    col = df.columns
    
    region_col = [c for c in col if c[:2] == region]
#    print(region_col)
    region_df = df[region_col]
    region_df.insert(0, '날짜', date)
#    print(region_df.head())
    return region_df
```

```python
## 2017년부터 자료 추출 

for i in range(len(df)):
    if df['날짜'][i] == '2016/01':
        df = df[i:]
df
```

```python
r_df = region_extract(df, '서울')
r_df.head()
```

```python
month_2016 = r_df[r_df['날짜'].between('2016/01', '2016/12')]
month_2017 = r_df[r_df['날짜'].between('2017/01', '2017/12')]
month_2018 = r_df[r_df['날짜'].between('2018/01', '2018/12')]
month_2019 = r_df[r_df['날짜'].between('2019/01', '2019/12')]
month_2020 = r_df[r_df['날짜'].between('2020/01', '2020/12')]
month_2021 = r_df[r_df['날짜'].between('2021/01', '2021/12')]
```

```python
month_2017.head()
```

```python
month_2016.set_index('날짜', inplace=True)
month_2017.set_index('날짜', inplace=True)
month_2018.set_index('날짜', inplace=True)
month_2019.set_index('날짜', inplace=True)
month_2020.set_index('날짜', inplace=True)
month_2021.set_index('날짜', inplace=True)
```

```python
month_2021
```

```python
x_2016 = month_2016.sum()
x_2017 = month_2017.sum()
x_2018 = month_2018.sum()
x_2019 = month_2019.sum()
x_2020 = month_2020.sum()
x_2021 = month_2021.sum()
```

```python
sum_df = pd.DataFrame(pd.concat([pd.DataFrame(x_2016).T, pd.DataFrame(x_2017).T, pd.DataFrame(x_2018).T, pd.DataFrame(x_2019).T, pd.DataFrame(x_2020).T, 
           pd.DataFrame(x_2021).T], axis=0))

sum_df
```

```python
sum_df['year'] = [2016,2017,2018,2019,2020,2021]
```

```python
sum_df=sum_df.set_index('year')
```

```python
sum_df
```

```python
def create_plot(df):
    col_names = df.columns
    col_names =[c.replace('/', "-") for c in col_names]
    df.columns = col_names
    
    for col in col_names:
        new_col = []
        analy_tgt = []

        for i in df.index[:-1]:
            n = (df[col][i+1] - df[col][i])/df[col][i] *100
            n = round(n, 2)
            new_col.append(n)
        for a in df[col].tolist()[1:]:
            
            a_leng = len(str(a)) -1
            a_leng *= '0'
            a_leng = int('1'+ a_leng)
            
            a = round(a/a_leng, 2)
            analy_tgt.append(a)
        
        x = df.index[1:]
        y1 = analy_tgt
        y2 = new_col
        fig, ax1 = plt.subplots()
        ax1.set_facecolor('white')
        

        ax1.bar(x, y1,  color='red', alpha=0.7, width=0.7)
        ax1.set_xlabel('Year')
        ax1.set_ylabel(f'{col}')
        ax1.set_ylim(ymin=int(min(y1))-1, ymax= int(max(y1))+1)
        ax1.tick_params(axis = 'both', direction ='in')

        ax2 = ax1.twinx()
        ax2.plot(x, y2, '-*', color='blue', markersize = 8, linewidth=2, alpha=0.7)
        ax2.set_ylabel('증감율')
        ax2.tick_params(axis ='y', direction='in')
        
        plt.savefig(f'./img/{col}_plot.png')
    return None
```

```python
create_plot(sum_df)
```
